# web scrapping for television using python scrapped prices,brands and ratings data from flipkart

```python
#!pip install requests_html
import pandas as pd 
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python

#scrAapping data from the url
p_name=[]
p_price=[]
p_rating=[]

url="https://www.flipkart.com/search?q=samsung+tv&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_10_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_10_na_na_na&as-pos=1&as-type=RECENT&suggestionId=samsung+tv%7CTelevisions&requestId=accb2fca-507b-4e11-9cc1-f3f9fe8f1e8d&as-searchtext=samsung%20te"
req=requests.get(url)
content=BeautifulSoup(req.content,"html.parser")
#print(content)
name=content.find_all("div",{"class":"_4rR01T"})
price=content.find_all("div",{"class":"_30jeq3 _1_WHN1"})
rating=content.find_all("div",{"class":"_3LWZlK"})
#print("phone in page"+str(i))
print((name))
for i in name:
    p_name.append(i.text)
for i in price:
    p_price.append(i.text)
for i in rating:
    p_rating.append(i.text)

    
    
```

```python
print("product_name",p_name)
len(p_name)
```

```python
print("product_price",p_price)
len(p_price)
```

```python
print("product_ratings",p_rating)
```

```python
#making length of every single variable same so that we can convert it into Dataframe
p_rating=p_rating[0:24]
len(p_rating)
```

```python
#to covert data from list into datafram
import pandas as pd
data={"Product_Name":p_name,"Product_Price":p_price,"Product_Rating":p_rating}
television=pd.DataFrame(data)
television
```
